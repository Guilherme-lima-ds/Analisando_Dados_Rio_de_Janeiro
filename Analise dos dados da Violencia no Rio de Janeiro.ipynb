### Etapa 2: Extrar e Limpar os dados.

```python
df = pd.read_csv('https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/violencia_rio.csv')
```

```python
#Primeiras entradas das informações
df.head()
```

- Como podemos ver, logo nas nossas 5 primeiras linhas temos valores nulos faltantes, vamos descobrir quanto são para tomamos a melhor decisão de como trata-los.

```python
#Verificando priemeiro a quantidade de linhas e colunas do Df
df.shape
```

```python
#Verificando a porcentagem de valores nulos das colunas
(df.isnull().sum()/df.shape[0]).sort_values(ascending=False)
```

- Como podemos notar, mais de 80% das colunas de furto de bicicleta e roub_bicicleta estão vazias, e como não podemos remover linhas em branco por causa da quantidade de linhas que temos no nossa Dataframe, vamos precisar exclui-las, pois tambem não podemos substituir mais de 80% dos dados, outro problema e que tem outras colunas com o mesmo problema, eu vou escolher não usaladas na minha analise, pois elas estão vindo assim da fonte.
- Qual vai ser meu criterio?
    - Colunas com mais de 40% dos valores nulos vão ser excluidas.

```python
#Verificando o nome das colunas
df.columns
```

```python
#Criando um novo df para retirar as colunas vazias
Clean_df = df.copy()

#Removendo colunas
Clean_df.drop(columns=['furto_bicicleta','roubo_bicicleta','cmba','posse_drogas','trafico_drogas','apreensao_drogas_sem_autor','apf','aaapai','cmp',
                'furto_celular','sequestro_relampago','extorsao','furto_coletivo','roubo_apos_saque','roubo_conducao_saque','roubo_cx_eletronico','estelionato','pol_militares_mortos_serv','estupro','pol_civis_mortos_serv'], 
              axis=1,inplace=True)
```

```python
#Verificando a porcentagem de valores ausentes de novo
(Clean_df.isnull().sum()/Clean_df.shape[0]).sort_values(ascending=False)
```

- As outras colunas que sobraram, eu vou substituir os valores pela mediana.

```python
#Substituindo o valor nulo de encontro ossada para a mediana dele
Clean_df.fillna(Clean_df['encontro_ossada'].median(),inplace=True)

#Substituindo os valores de pessoas_desaparecidas pela mediana
Clean_df.fillna(Clean_df['pessoas_desaparecidas'].median(),inplace=True)

#Substituindo os valores de roubo_celular pela mediana
Clean_df.fillna(Clean_df['roubo_celular'].median(),inplace=True)

#Substituindo os valores de ameaca pela meadiana
Clean_df.fillna(Clean_df['ameaca'].median(),inplace=True)

#Substituindo os valores de registro_ocorrencias pela mediana
Clean_df.fillna(Clean_df['registro_ocorrencias'].median(),inplace=True)

#Subsitindo os valores de lesao_corp_morte pela mediana
Clean_df.fillna(Clean_df['lesao_corp_morte'].median(),inplace=True)

#Substituindo os valores de sequestro pela mediana
Clean_df.fillna(Clean_df['sequestro'].median(),inplace=True)

#Substitindo os valores de lesao_corp_culposa pela mediana
Clean_df.fillna(Clean_df['lesao_corp_culposa'].median(),inplace=True)

#Substituindo os valores de hom_por_interv_policial pela mediana
Clean_df.fillna(Clean_df['hom_por_interv_policial'].median(),inplace=True)

#Substituindo os valores de hom_culposo pela mediana
Clean_df.fillna(Clean_df['hom_culposo'].median(),inplace=True)
```

```python
#Verificando a porcentagem de valores ausentes de novo
(Clean_df.isnull().sum()/Clean_df.shape[0]).sort_values(ascending=False)
```

### Etapa 3: Analisar os dados.

- E a média de furtos de carros qual é?

```python
#Verificando e arrendondando a media
round(Clean_df['furto_veiculos'].mean(),2)
```

- Qual a media de recuperação de veiculos?

- Qual a porcentagem de carros recuperados em relação aos carros roubados + furtados?

```python
porc_veiculo = Clean_df['recuperacao_veiculos'].mean() / (Clean_df['furto_veiculos'].mean() + Clean_df['roubo_veiculo'].mean())
print('A taxa de recuperação dos veiculos versus os carros roupados e de {:.2%}'.format(porc_veiculo))
```

- Metricas de violencia por mes

```python
#Criando uma função para criar os graficos das metricas por mes

def Metricasviolencia(coluna):
    mes_group = Clean_df.groupby('mes')[coluna].sum()
    return sns.lineplot(x=mes_group.index, y=mes_group)
```

```python
#Usando a função para saber a soma das ocorrencia de latriocinio por mes
Metricasviolencia('latrocinio');
```

```python
#Soma de pessoas desaparecidas por mes
Metricasviolencia('pessoas_desaparecidas');
```

```python
#Roubo a banco por mes
Metricasviolencia('roubo_banco');
```

```python
#Roubo de celular por mes
Metricasviolencia('roubo_celular');
```

```python
#Sequesto por mes
Metricasviolencia('sequestro');
```

- Analise geral dos graficos: E interresante notar como a maiorias dos graficos tem picos e quedas em certo meses como latricio e pessoas desaparecidas, pode ser pelo fato do assaltante ter disparado acidentalmente ou por que quis mesmo e escondeu o corpo, e por isso há o aumento nessa epoca de pessoas desaparecidas pois o latrocio(Roubo a mão armada) aumenta tambem.  
---
Lembrando que foi dessa coluna que eu substitui os valores nulos pela mediada, pode ter algo relacionado tambem pois naquele mes não teve pessoas desaparecidas ou esquecerão de registrar, por isso trabalhar com dados ausentes e complicado, mas podemos deixar assim ou excluir a coluna, ai vai da nossa escolha.

```python
#Criando uma função para criar os graficos das metricas por ano

def Metricasviolencia_ano(coluna):
    mes_group = Clean_df.groupby('vano')[coluna].sum()
    return sns.lineplot(x=mes_group.index, y=mes_group)
```

```python
#Soma de pessoas desaparecidas por ano
Metricasviolencia_ano('pessoas_desaparecidas');
```

- Aqui e legal, pois não podemos considerar que antes de 2000 mil não tinha pessoas desaparecidas, o que não tinha era registro dessas pessoas, tanto que essa coluna que eu preenchi com a mediana, deveria ser algum valor depois de 2000 mil para dar essas curvas assim.

```python
#Soma de hom_doloso por ano
Metricasviolencia_ano('hom_doloso');
```

- Interresante e notar como cai em 2020, bem no ano da pandemia será que foi por que não tinha mais ou não tinha ninguem para registrar? esse e o problema dos dados publicos, as vezes vem com muito erro de informação e dados nulos.

```python
#Soma de hom_por_interv_policial por ano
Metricasviolencia_ano('hom_por_interv_policial');
```
